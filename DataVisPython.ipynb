## BoxPlot

Five-number summary:

1.   min = minimum value
2.   25% = first quartile (Q1) = median of the lower half of the data
3.   50% = second quartile (Q2) = median of the data
4.   75% = third quartile (Q3) = median of the upper half of the data
5.   max = maximum value

```
x1 = np.random.normal(None)
x2 = np.random.normal(None)
x3 = np.random.normal(None)
x4 = np.random.normal(None)
data = None

plt.boxplot(None, 
            notch=False, # box instead of notch shape 
            sym='rs',    # red squares for outliers
            vert=True,   # vertical box aligmnent
            labels = ['x1', 'x2', 'x3', 'x4'])

plt.xlabel('Data')
plt.title('Box plot')
```
