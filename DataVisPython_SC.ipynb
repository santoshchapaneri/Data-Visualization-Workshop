The Gaussian PDF is given by

$X \sim \mathcal{N}(\mu, \sigma^2) \implies p(x) = \frac{1}{\sqrt{2 \pi \sigma^2}} \exp\left(- \frac{1}{2 \sigma^2} (x - \mu)^2 \right) $


```
data = np.random.normal(0, 20, 1000)    # mean, std, numSamples 

# fixed bin size
mybins = np.arange(-100, 100, 5)
plt.xlim([min(data)-5, max(data)+5])
plt.hist(data, bins=mybins)
plt.title('Random Gaussian data')
plt.xlabel('X')
plt.ylabel('Frequency')
plt.show()
```

```
# Histogram of 2 overlapping Gaussians

data1 = np.random.normal(15, 5, 1000)    # mean, std, numSamples 
data2 = np.random.normal(-15, 10, 1000)    # mean, std, numSamples 

mybins = np.arange(-60, 60, 5)
plt.xlim([min(data1+data2)-5, max(data1+data2)+5])

plt.hist(data1, bins=mybins, alpha=0.75, label='Class 1')    
plt.hist(data2, bins=mybins, alpha=0.75, label='Class 2')

plt.title('Random Gaussian data')
plt.xlabel('X')
plt.ylabel('Frequency')
plt.legend(loc='upper right')
```

```
# creating random vectors from the multivariate normal distribution
mu_vec1 = np.array([0,0])
cov_mat1 = np.array([[2,0],[0,2]])

x1_samples = np.random.multivariate_normal(mu_vec1, cov_mat1, 1000)
x2_samples = np.random.multivariate_normal(mu_vec1+5, cov_mat1+0.8, 1000)

# print(x1_samples.shape)
# (100, 2), 100 rows, 2 columns

plt.figure(figsize=(8,6))
plt.scatter(x1_samples[:,0], x1_samples[:,1], marker='x', 
            color='blue', alpha=0.7, label='x1 samples')
plt.scatter(x2_samples[:,0], x2_samples[:,1], marker='o', 
            color='green', alpha=0.7, label='x2 samples')
plt.title('Basic scatter plot')
plt.ylabel('X')
plt.xlabel('Y')
plt.legend(loc='upper right')
plt.show()
```

## BoxPlot

Five-number summary:

1.   min = minimum value
2.   25% = first quartile (Q1) = median of the lower half of the data
3.   50% = second quartile (Q2) = median of the data
4.   75% = third quartile (Q3) = median of the upper half of the data
5.   max = maximum value

```
x1 = np.random.normal(100,10,500)
x2 = np.random.normal(80,30,500)
x3 = np.random.normal(90,20,500)
x4 = np.random.normal(70,60,500)
data = [x1, x2, x3, x4]

plt.boxplot(data, 
            notch=False, # box instead of notch shape 
            sym='rs',    # red squares for outliers
            vert=True,   # vertical box aligmnent
            labels = ['x1', 'x2', 'x3', 'x4'])

plt.xlabel('Data')
plt.title('Box plot')
```

```
import pandas as pd

from google.colab import files
uploaded = files.upload()
```

```
import io

drink_cols = ['country', 'beer', 'spirit', 'wine', 'liters', 'continent']

drinks = pd.read_csv(io.BytesIO(uploaded['drinks.csv']),header=0, names=drink_cols)
```

## Faceted histograms
Viewing data via histograms of subsets.

**Tips dataset**: shows the amount that the restaurant staff receive in tips

```
tips = sns.load_dataset('tips')
tips.head()
```

```
tips.shape
```

```
# Creating a new column called tip_pct (percentage of total bill)
 
tips['tip_pct'] = 100 * tips['tip'] / tips['total_bill']

tips.head()
```

```
grid = sns.FacetGrid(tips, row = "sex", col = "time", margin_titles = True)
grid.map(plt.hist, "tip_pct", bins = np.linspace(0, 40, 15))
```

## Factor plots
Factor plots allows to view the distribution of a parameter within bins defined by any other parameter.

```
g = sns.factorplot("day", "total_bill", "sex", data = tips, kind = "box")
g.set_axis_labels("Day", "Total Bill")
```

```
# Joint distribution
sns.jointplot("total_bill", "tip", data = tips, kind='kde')
```
